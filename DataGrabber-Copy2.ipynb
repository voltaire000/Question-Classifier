```python
import pandas as pd
import numpy as np
import requests
import datetime as time
```

```python
url = 'https://api.pushshift.io/reddit/submission/search'
dfs = []
loops = 2000
subreddit = 'amitheasshole'
params = {
    'subreddit': subreddit,
    'size': 100,
}
start_time = time.time()
for i in range(loops):
    current_time = time.time()
    #requesting data
    try:
        res = requests.get(url, params)
        print(f'res {i+1} code: ', res.status_code)
        data = res.json()
        posts = data['data']
        dfs.append(pd.DataFrame(posts))
        #get oldest post time and use as before parameter in next request
        oldest = posts['created_utc'].min()
        params['before'] = oldest
        #sleep for .5 second
        time.sleep(.5)
    except:
        pass
reddit_posts = pd.concat(dfs)
filename = subreddit + '.csv' # thank you hov for this code to get the reddit posts
```

###### 

```python
reddit_posts.to_csv('./datasets/' + filename, index=False) # moving the data from amitheasshole to a dataframe
```
